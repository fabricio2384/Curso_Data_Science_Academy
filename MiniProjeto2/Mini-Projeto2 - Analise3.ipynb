## Análise 3

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

```python
df.T
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
Fuel, ax = plt.subplots(figsize=(15,10))
sns.barplot(x='fuelType', y='price', hue='gearbox', data=df,palette='GnBu_r',errwidth=1,capsize=.2 )
sns.set_style('white')
ax.xaxis.set_label_text('Tipo de Veiculo',fontdict= {'size':20})
ax.yaxis.set_label_text('Variação do preço de venda x1000',fontdict= {'size':20})
ax.ticklabel_format(axis='y',style='sci', scilimits=(3,3))
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# Salvando o plot
Fuel.savefig("plots/Analise3/Combustivel_Preço.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
Power, ax = plt.subplots(figsize=(15,10))
sns.barplot(x='vehicleType', y='powerPS', hue='gearbox', data=df,palette='Blues',errwidth=1,capsize=.2 )
sns.set_style('white')
ax.xaxis.set_label_text('Tipo de Veiculo',fontdict= {'size':20})
ax.yaxis.set_label_text('Variação do preço de venda x1000',fontdict= {'size':20})
ax.ticklabel_format(axis='y',style='sci', scilimits=(3,3))
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# Salvando o plot
Power.savefig("plots/Analise3/Power_Tipo_Veiculo.png")
```
