## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
#import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

```python
df.head()
```

```python
df['yearOfRegistration'].count()
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

his, ax = plt.subplots(figsize=(8,6))
sns.histplot(df['yearOfRegistration'],bins=50, color='#235678',kde=True, ax=ax)
ax.set_title('Distribuição de Veículos com base no Ano de Registro', fontsize= 15)
plt.ylabel("Densidade (KDE)", fontsize= 15)
plt.xlabel("Ano de Registro", fontsize= 15)
plt.show()
 
```

```python
# Salvando o plot
his.savefig("plots/Analise1/Distri_Veiculos x Ano de Registro.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers
box, ax = plt.subplots(figsize=(15,10))
sns.set_style('white')

sns.boxplot(x='vehicleType', y='price', data=df, palette='Blues')
ax.text(5.25,27000,'Análise de Outliers',fontsize=25,color='blue',alpha=0.5, ha='center', va='center')
ax.xaxis.set_label_text('Tipo de Veiculo',fontdict= {'size':20})
ax.yaxis.set_label_text('Variação do preço de venda x1000',fontdict= {'size':20})
ax.ticklabel_format(axis='y',style='sci', scilimits=(3,3))
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)



plt.show()
```

```python
# Salvando o plot
box.savefig("plots/Analise1/Preço por veiculo.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
count, ax = plt.subplots(figsize=(15,10))

sns.set_style('white')
plt.title('Contagem total de veículos à venda conforme o tipo de veículo',fontsize=16,style='oblique')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ticklabel_format(axis='y',style='sci', scilimits=(3,3))
plt.xlabel('Tipo de Veiculo',fontsize=20)
plt.ylabel('Distribuição de Veiculos Para Vendas (x1000)',fontsize=20)
ax = sns.countplot(x="vehicleType", data=df, palette='light:b')
for i in ax.patches:
    ax.annotate(i.get_height(), 
                        (i.get_x() + i.get_width() / 2.0
                         ,i.get_height()
                        )
                        ,ha = 'center'
                        ,va = 'center'
                        ,xytext = (0, 5)
                        ,textcoords = 'offset points'
                        )
```

```python
# Salvando o plot
count.savefig("plots/Analise1/Contagem de veiculos.png")
```
